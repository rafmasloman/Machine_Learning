```python
import numpy as numpy
import pandas as panda
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
ds = panda.read_csv("bread basket.csv") #membaca file csv
ds.head() #menampilkan data teratas
```

```python
ds=ds.drop(ds.loc[ds['Item']=='NONE'].index) #menghapus kolom Item pada index yang bernilai None 
ds.groupby('period_day')['Item'].count().sort_values(ascending=False) #mengurutkan jumlah banyaknya transaksi yang terjadi pada tabel period_day 
```

```python
lst=[]
for item in ds['Transaction'].unique():
    lst2=list(set(ds[ds['Transaction']==item]['Item']))
    if len(lst2)>0:
        lst.append(lst2)
print(lst[0:3])
print(len(lst))
```

```python
te=TransactionEncoder()
te_data=te.fit(lst).transform(lst)
data_x=panda.DataFrame(te_data,columns=te.columns_)
data_x.head()
```

```python
#Algoritma Apriori
frequent_items= apriori(data_x, use_colnames=True, min_support=0.03)
print(frequent_items.head())

rules = association_rules(frequent_items, metric="lift", min_threshold=1)
rules.antecedents = rules.antecedents.apply(lambda x: next(iter(x)))
rules.consequents = rules.consequents.apply(lambda x: next(iter(x)))
rules
```
